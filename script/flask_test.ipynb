```python
import sys
import sympy

def Fib(n):
    x = sympy.symbols('x', nonnegative=True, integer=True)
    Fib = 1 / sympy.sqrt(5) * (((1+sympy.sqrt(5))/2)**(n-1) - ((1-sympy.sqrt(5))/2)**(n-1))
    result = Fib.subs(x, n)
    result = sympy.simplify(result)
    return result

def main(lines):
    fib_list = []
    N = int(lines[0])

    for i in range(10):
        fib_list.append(Fib(i))

    if N in fib_list:
        print(fib_list.index(N)-1)
    else:
        print(-1)

if __name__ == '__main__':
    lines = []
    for l in sys.stdin:
        lines.append(l.rstrip('\r\n'))
        main(lines)
```
